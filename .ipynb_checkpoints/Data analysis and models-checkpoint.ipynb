# Load the Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv('datasets/train.csv')
df.head()
```

# Data Cleaning

```python
df.isnull().sum().sort_values(ascending=False)[:30]
```

```python
categ_columns = ['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu', 'Garage Finish', 'Garage Cond', 
                 'Garage Qual', 'Garage Type', 'Bsmt Exposure', 'BsmtFin Type 2', 'BsmtFin Type 2', 'Bsmt Cond', 
                 'Bsmt Qual', 'Mas Vnr Type']

[df[x].replace(np.nan, 'NA', inplace=True) for x in categ_columns]

df.head()
```

```python
num_columns = ['Lot Frontage', 'Garage Yr Blt', 'Mas Vnr Area', 'Bsmt Half Bath', 'Bsmt Full Bath', 
                'Garage Cars', 'Garage Area', 'Bsmt Unf SF', 'BsmtFin SF 2', 'Total Bsmt SF', 'BsmtFin SF 1']

[df[x].replace(np.nan, '0', inplace=True) for x in num_columns]

df.head()
```

```python
df.replace('None', 'NA', inplace=True)
df['Mas Vnr Type'].unique()
```

```python
df.info()
```

```python
# change from object to numbers

float = ['Lot Frontage', 'Mas Vnr Area', 'Garage Cars', 'Garage Area']

for x in float:
    df[x] = df[x].astype('float64', copy=False)
```

```python
# get dummies

df = pd.get_dummies(df, columns = ['MS Zoning', 'Street', 'Land Contour', 'Lot Config', 'Land Slope', 'Bldg Type', 
                    'Paved Drive'], drop_first=True)
```

```python
# Transform categorical variables to ordinal

transform = ['Exter Qual', 'Exter Cond', 'Bsmt Qual', 'Bsmt Cond', 'Heating QC', 'Kitchen Qual', 'Fireplace Qu',
             'Garage Qual', 'Garage Cond', 'Pool QC']

rankings = {'NA':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5}

for col in transform:
    df[col] = df[col].map(lambda col: rankings.get(col, col))
```

```python
# Total square feet

sqft = [x for x in df.columns if 'SF' in x]

df['Total square ft'] = df[sqft].sum(axis=1)
```

```python
# Total bathrooms

baths = [x for x in df.columns if 'Bath' in x]
df['Total baths'] = df[baths].sum(axis=1)
df.drop(['Full Bath', 'Half Bath', 'Bsmt Full Bath', 'Bsmt Half Bath'], axis=1, inplace=True)
```

```python
# If house has the feature

# Central Air
df['Has Central Air'] = np.where(df['Central Air'].str.contains('N'), 0, 1)
df.drop(['Central Air'], axis=1, inplace=True)
```

```python
# Garage
df['Has Garage'] = np.where(df['Garage Type'].str.contains('NA'), 0, 1)
df.drop(['Garage Type'], axis=1, inplace=True)
```

```python
# Fence
df['Has Fence'] = np.where(df['Fence'].str.contains('NA'), 0, 1)
df.drop(['Fence'], axis=1, inplace=True)
```

```python
# Age
from datetime import datetime
now = datetime.now()
df['Current Year'] = now.year
df['Age'] = df['Current Year'] - df['Year Built']
df.drop(['Current Year', 'Year Built', 'Mo Sold', 'Yr Sold', 'Misc Feature'], axis=1, inplace=True)
```

```python
# Drop the remaining onject columns

df.drop(df.select_dtypes(include=['object']).columns, axis=1, inplace=True)
```

```python
df.head()
```

```python
# Save the clean dataset

df.to_csv('./datasets/train_clean.csv', index=False)
```

# Modeling

```python
# Heatmap
plt.figure(figsize=(8,14))
sns.heatmap(df.corr()[['SalePrice']].sort_values('SalePrice', ascending=False),
            annot=True, cmap='coolwarm', vmin=-1, vmax=1)
```

```python
# Select variables whose correlation with Sale Price is positive

variables = df.corr()['SalePrice'][df.corr()['SalePrice'] > 0]
variables = variables.drop('SalePrice')
variable_list = [x for x in variables.index]
variable_list
```

```python
X = df[variable_list]
y = df['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

Since I predict the price of a house, and I have features are on different scales (ex. square feet vs number of bathroom). I scale our data here.

```python
sc = StandardScaler()
sc.fit(X_train)
```

```python
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

### LinearRegression

```python
lr = LinearRegression()
lr.fit(X_train_sc, y_train)
```

```python
lr.score(X_train_sc, y_train), lr.score(X_test_sc, y_test), cross_val_score(lr, X_train_sc, y_train).mean()
```

### Ridge

```python
ridge = RidgeCV()
ridge.fit(X_train_sc, y_train)
```

```python
ridge.score(X_train_sc, y_train), ridge.score(X_test_sc, y_test), cross_val_score(ridge, X_train_sc, y_train).mean()
```

### LASSO

```python
lasso = LassoCV()
lasso.fit(X_train_sc, y_train)
```

```python
lasso.score(X_train_sc, y_train), lasso.score(X_test_sc, y_test), cross_val_score(lasso, X_train_sc, y_train).mean()
```

From the r2 scoring above, I find that RidgeCV has the highest score on training and cross validation while the difference between training and testing is lower than LASSO. I decide to run the Ridge model on the unseen data.

```python
r_preds = ridge.predict(X_test_sc)
```

```python
r_residuals = y_test - r_preds
```

```python
plt.figure(figsize=(10,6))
plt.scatter(r_preds, r_residuals)
plt.axhline(y=0, color='r')
plt.title('Residual Plot', fontsize=16)
plt.xlabel('predictions')
plt.ylabel('residuals');
```

Residual plot shows the errors corresponding to the predicted values is randomly distributed. It looks normal, so I can go ahead and use the model.
